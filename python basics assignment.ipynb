```python
# 4. Discuss the different types of operators in python and provide examples of how they are used.
# ans.- Python supports several types of operators, each serving different purposes. Hereâ€™s a breakdown of the main types of operators with examples:
       #1. Arithmetic Operators
       # These operators perform mathematical operations.

# Addition (+)
# Subtraction (-)
# Multiplication (*)
# Division (/)
# Floor Division (//)
# Modulus (%)
# Exponentiation (**)

#Example:-
a = 10
b = 3

print("Addition:", a + b)         # 13
print("Subtraction:", a - b)      # 7
print("Multiplication:", a * b)   # 30
print("Division:", a / b)         # 3.3333...
print("Floor Division:", a // b)   # 3
print("Modulus:", a % b)          # 1
print("Exponentiation:", a ** b)   # 1000

      # 2. Comparison Operators
      # These operators compare two values and return a Boolean result (True or False).

# Equal (==)
# Not Equal (!=)
# Greater Than (>)
# Less Than (<)
# Greater Than or Equal To (>=)
# Less Than or Equal To (<=)

#Example:-
x = 5
y = 10

print("Equal:", x == y)               # False
print("Not Equal:", x != y)           # True
print("Greater Than:", x > y)         # False
print("Less Than:", x < y)            # True
print("Greater Than or Equal To:", x >= y)  # False
print("Less Than or Equal To:", x <= y)      # True

       # 3. Logical Operators
       # These operators are used to combine conditional statements.

# AND (and)
# OR (or)
# NOT (not)

#Example:-
a = True
b = False

print("AND:", a and b)      # False
print("OR:", a or b)        # True
print("NOT:", not a)        # False

        # 4. Assignment Operators
        # These operators assign values to variables, often using an arithmetic operation.

# Assign (=)
# Add and Assign (+=)
# Subtract and Assign (-=)
# Multiply and Assign (*=)
# Divide and Assign (/=)

#Example:-
num = 5
num += 3      # Equivalent to num = num + 3
print(num)    # 8

num *= 2      # Equivalent to num = num * 2
print(num)    # 16

        # 5. Bitwise Operators
        # These operators perform bit-level operations on integers.

# AND (&)
# OR (|)
# XOR (^)
# NOT (~)
# Left Shift (<<)
# Right Shift (>>)

#Example:-
a = 5   # Binary: 0101
b = 3   # Binary: 0011

print("AND:", a & b)        # 1 (0001)
print("OR:", a | b)         # 7 (0111)
print("XOR:", a ^ b)        # 6 (0110)
print("NOT:", ~a)           # -6 (Inverts bits)
print("Left Shift:", a << 1)  # 10 (1010)
print("Right Shift:", a >> 1) # 2 (0010)

        # 6. Identity Operators
        # These operators check if two variables point to the same object in memory.

# Is (is)
# Is Not (is not)

#Example:-
list1 = [1, 2, 3]
list2 = list1
list3 = list1.copy()

print("list1 is list2:", list1 is list2)       # True
print("list1 is list3:", list1 is list3)       # False
print("list1 is not list3:", list1 is not list3)  # True

        #7. Membership Operators
        # These operators check for membership in sequences (like strings, lists, or tuples).

# In (in)
# Not In (not in)

#Example
my_list = [1, 2, 3, 4, 5]

print("3 in my_list:", 3 in my_list)          # True
print("6 not in my_list:", 6 not in my_list)  # True

        #Summary
        # Operators in Python allow for a wide range of operations, from arithmetic to logical evaluations. 
        # these operators is essential for effective programming and data manipulation in Python.
```

```python
# 5. Explain the concept of type casting in python with example.

# ans.- Type casting in Python refers to the process of converting a variable from one data type to another.
        # This can be useful when you need to perform operations that require specific data types or when you want to ensure that data is in the right format.
        # Types of Casting
       #1.Implicit Casting: Python automatically converts one data type to another without explicit instruction.
        # This usually happens in expressions with mixed types.

# Example:-

a = 10        # int
b = 5.5      # float
c = a + b    # a is implicitly cast to float
print(c)     # Output: 15.5

       # 2.Explicit Casting: You can explicitly convert a variable from one type to another using built-in functions like int(), float(), str(), etc.

# Example:-
# Convert float to int:
num_float = 10.7
num_int = int(num_float)  # This will truncate the decimal
print(num_int)            # Output: 10

# Convert int to float:
num_int = 10
num_float = float(num_int)
print(num_float)          # Output: 10.0

# Convert int to string:
num_int = 10
num_str = str(num_int)
print(num_str)           # Output: '10'

# Convert string to int (note that the string must represent a valid integer):
num_str = "20"
num_int = int(num_str)
print(num_int)           # Output: 20

        # Summary
        # Type casting is a useful feature in Python that allows for flexible manipulation of different data types, making it easier to perform calculations
        # and work with data. explicit casting should be used when necessary, especially when dealing with user inputs or string representations of numbers.
```

```python
# 6. How do conditional statements work in python? illustrate with example

# ans.-   Conditional statements in Python allow you to execute certain blocks of code based on whether a condition is true or false.
        # The primary conditional statements are if, elif, and else.
# Syntax

if condition:
    # Code to execute if condition is true
elif another_condition:
    # Code to execute if another_condition is true
else:
    # Code to execute if none of the above conditions are true
    
# Example:-
age = 20

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
# 7. Describe the different types of loops in python and their use cases with examples.

# ans.- In Python, loops are used to execute a block of code repeatedly. The two main types of loops are for loops and while loops.
        # Here's an overview of each type along with their use cases and examples:
        # 1. for Loop:
        # The for loop is used to iterate over a sequence (like a list, tuple, string, or range). 
        # It is ideal for situations where you know in advance how many times you want to execute a block of code.
        
        # Use Cases:
        # Iterating through a collection of items.
        # Performing an action a specific number of times.
        
# Example:-
# Iterating through a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# Using range to iterate a specific number of times
for i in range(5):
    print(i)
    
        # 2. while Loop:
        # The while loop continues to execute as long as a specified condition is true. Itâ€™s useful when the number of iterations is not known beforehand.
        # Use Cases:
        # Continuously checking for a condition (e.g., user input).
        # Situations where the loop might need to run indefinitely until a specific condition is met
        
# Example:-
# A simple countdown
count = 5
while count > 0:
    print(count)
    count -= 1
    
        # 3. Nested Loops:
        # You can also nest loops within each other. This is useful for multi-dimensional data structures 
        # (like lists of lists) or when you need to perform combinations of iterations.

        # Use Cases:
        # Iterating over a matrix or grid.
        # Generating combinations of items
        
# Example:-
# Nested loops to print a multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i * j}")
        
        # 4. Loop Control Statements:
        # You can use break, continue, and pass to control the flow of loops:

        # break: Exits the loop prematurely.
        # continue: Skips the current iteration and moves to the next.
        # pass: A placeholder that does nothing; it can be used when a statement is syntactically required.
        
#Example:-
# Using break to exit a loop
for num in range(10):
    if num == 5:
        break
    print(num)

# Using continue to skip even numbers
for num in range(10):
    if num % 2 == 0:
        continue
    print(num)  # Will print only odd numbers
    
        # Conclusion:
        # Loops in Python provide powerful mechanisms for iteration, allow to perform repetitive tasks efficiently. 
        # The choice between for and while loops depends on whether to know the number of iterations in advance or if need to keep looping until
        # a condition changes. Nested loops can handle complex data structures, while control statements help manage loop execution flow.
```
